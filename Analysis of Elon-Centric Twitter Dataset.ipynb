## List of Functions Created and Utilized

```python
#Function 1: input username and search all tweets by user for a specific word
def get_text_tweets(textoftweet, screen_name):
   
    usertweets = stats_coll.find({'user.screen_name': screen_name,'text': {'$regex': textoftweet, '$options': 'i'}})
    
    i=0
    list = []
    for stats in usertweets:
        #print(stats["text"], stats["created_at"])
        list.append(stats)
        i=i+1
    print(i)
    
    return
 
#Testing
print("Function 1: Searching tweets for a keyword""")
print("How many times has Elon Musk tweeted about SpaceX?")
get_text_tweets('spacex', 'elonmusk')
print('How many times has Elon Musk tweeted about Tesla?')
get_text_tweets('Tesla', 'elonmusk')
print('How many times has Elon Musk tweeted about stocks?')
get_text_tweets('stock', 'elonmusk')
print('How many times has Elon Musk tweeted about Bitcoin?')
get_text_tweets('Bitcoin', 'elonmusk')


print('')
print('How many times has SpaceX tweeted about space?')
get_text_tweets('Space', 'SpaceX')
print('How many times has SpaceX tweeted about Elon Musk?')
get_text_tweets('elon', 'SpaceX')
print('How many times has SpaceX tweeted about rockets?')
get_text_tweets('rocket', 'SpaceX')
print('How many times has SpaceX tweeted about (rocket) launches?')
get_text_tweets('launch', 'SpaceX')
print('')
print('How many times has Tesla tweeted about SpaceX?')
get_text_tweets('SpaceX', 'Tesla')
print('How many times has Tesla tweeted about their Model car?')
get_text_tweets('model', 'Tesla')
print('How many times has Tesla tweeted about Elon Musk?')
get_text_tweets('elon', 'Tesla')
get_text_tweets('musk', 'Tesla')
print('')
print('')
```

```python
#Function 2: input username and it will return the most favorited tweet of the user
def get_most_favorited(screen_name):
    favoritedtweet = favor_coll.find({'user.screen_name': screen_name}).sort([('favorite_count', 1)]).limit(1)
    print(favoritedtweet[0])

#Testing
print("Function 2: Finding users most favorited tweet")
print("")
print("Elon Musk's most favorited tweet is ")
get_most_favorited('elonmusk')
print("")
print("SpaceX's most favorited tweet is ")
get_most_favorited('SpaceX')
print("")
print("Tesla's most favorited tweet is ")
get_most_favorited('Tesla')
print("")
print("")
print("")
```

```python
#Function 3: input username and it will return the most retweeted tweet 
def get_most_retweeted(screen_name):
    retweetedtweet = retweets_coll.find({'retweeted_status.user.screen_name': screen_name}).sort([('retweeted_status.retweet_count', 1)]).limit(1)
    print(retweetedtweet[0])

#Testing:
print("Function 3: Finding users most retweeted tweet")
print("")
print("Elon Musk's most retweeted tweet is ")
get_most_retweeted('elonmusk')
print("")
print("SpaceX's most retweeted tweet is ")
get_most_retweeted('SpaceX')
print("")
print("Tesla's most retweeted tweet is ")
get_most_retweeted('Tesla')
print("")
print("")
print("")
```

```python
#Function 4: input start and enddate or username and it will return the average for a 7 day period
def get_weekly_average(screen_name, startdate, enddate):
    date_time_startdate = datetime.strptime(startdate, '%Y-%m-%d')
    date_time_enddate = datetime.strptime(enddate, '%Y-%m-%d')
    
    min_date = date_time_startdate
    max_date = date_time_startdate + timedelta(days = 7)
    
    i = 0
    max_start_date = None
    max_end_date = None
    max_average = 0
    while (min_date <= date_time_enddate):
        print("This week's range was", min_date, max_date)
        usertweets = stats_coll.count_documents({"user.screen_name": screen_name,"created_at": {"$gte": min_date, "$lte": max_date}})
        print("Average tweets per week", usertweets/7)
        print("Total number of tweets this week:", usertweets)
        print("")
        if(usertweets/7 >= max_average):
            max_average = usertweets/7
            max_start_date = min_date
            max_end_date = max_date
        min_date = max_date
        max_date = max_date + timedelta(days = 7)
        
    usertweets = stats_coll.find({"user.screen_name": screen_name,"created_at": {"$gte": max_start_date, "$lte": max_end_date}})
   
    i=0
    list = []
    for stats in usertweets:
        #print(stats["text"],stats["created_at"])
        list.append(stats)
        i=i+1
    print("Greatest number of tweets per week was", i)
    print("This week's range was", max_start_date, max_end_date)
    
    return

#Testing    
print("Function 4: Finding the number of tweets per week and their averages")
print("")
print("Averages of Elon's Tweets from January 1, 2020 to December 31, 2020")
print(get_weekly_average('elonmusk', '2020-01-01', '2020-12-31'))
print("")
print("")

print("Averages of SpaceX's Tweets from January 1, 2020 to December 31, 2020")
print(get_weekly_average('SpaceX', '2020-01-01', '2020-12-31'))
print("")
print("")
print("Averages of Tesla's Tweets from January 1, 2020 to December 31, 2020")
print(get_weekly_average('Tesla', '2020-01-01', '2020-12-31'))
print("")
print("")
print("")
```

```python
#Function 5: input username and it will find the user that has mentioned them the most
def get_most_mentioneduser(screen_name):
    mentioneduser = retweets_coll.aggregate(
        [
            {
                '$match':{"user_mentions.screen_name": screen_name}
            },
            {
                '$group': {
                    '_id' : '$user.screen_name', 
                    'count' : {
                        '$sum': 1
                    }
                }
            },
            {
                '$sort': {'count': -1}
            }
        ]
    )
    
    print(list(mentioneduser)[0])
    


#Test:
print("Function 5: Finding the users who has mention the specified user the most")
print("")
print("Elon Musk was most mentioned by:")
get_most_mentioneduser('elonmusk')
print('')
print("Tesla was most mentioned by:")
get_most_mentioneduser('Tesla')
print('')
print("SpaceX was most mentioned by:")
get_most_mentioneduser('SpaceX')
print('')
print('')
```
